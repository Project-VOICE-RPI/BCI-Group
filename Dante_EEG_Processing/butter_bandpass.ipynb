```python
import numpy as np
import scipy.signal as signal
import matplotlib.pyplot as plt
```

```python
fs = 500.0
lowcut = 50.0
highcut = 150.0
order = 4

nyq = 0.5 * fs
low = lowcut / nyq
high = highcut / nyq

b, a = signal.butter(order, [low, high], btype='band')
w, h = signal.freqz(b, a, worN=2000)
```

```python
plt.plot((fs * 0.5 / np.pi) * w, abs(h))
plt.title("Butterworth filter frequency response")
plt.xlabel("Frequency [Hz]")
plt.ylabel("Gain")
plt.grid()
plt.show()
```

```python
t = np.linspace(0, 1, int(fs), endpoint=False)
signal_data = np.sin(2 * np.pi * 100.0 * t) + 0.5 * np.sin(2 * np.pi * 200.0 * t)

filtered_data = signal.filtfilt(b, a, signal_data)
```

```python
plt.figure()
plt.plot(t, signal_data, label="Original Signal")
plt.plot(t, filtered_data, label="Filtered Signal", linewidth=2)
plt.legend()
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Signal Before and After Butterworth Bandpass Filter")
plt.grid()
plt.show()
```
